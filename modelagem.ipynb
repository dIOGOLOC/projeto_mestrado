```python
fig, ax = plt.subplots()
fig.patch.set_facecolor('white')
for i in xrange(len(STdados['st'])):
    ax.plot(STdados['time'][i]-5,STdados['data'][i]*10+dist_perfil1[i], 'k', linewidth=1)
    ax.fill_between(STdados['time'][i]-5,STdados['data'][i]*10+dist_perfil1[i],dist_perfil1[i], where = STdados['data'][i]*10+dist_perfil1[i]>=dist_perfil1[i], facecolor='black',alpha=0.8, interpolate=True)
    ax.fill_between(STdados['time'][i]-5,STdados['data'][i]*10+dist_perfil1[i],dist_perfil1[i], where = STdados['data'][i]*10+dist_perfil1[i]<=dist_perfil1[i], facecolor='gray',alpha=0.5, interpolate=True)
    ax.set_xlim(-2,20)
    ax.text(-3,dist_perfil1[i]+0.5,perfil1[i],rotation=90,fontsize=8)
    #plt.text(21,140,'Distance (km)',rotation=90,fontsize=20)
    ax.tick_params(axis='y', labelleft='off', labelright='off',direction=180,pad=10)
    ax.spines['bottom'].set_bounds(-2, 20)
    ax.spines['left'].set_bounds(0, 10)
    # Hide the right and top spines
    ax.spines['right'].set_visible(False)
    ax.spines['top'].set_visible(False)
    ax.xaxis.set_ticks_position('bottom')
    plt.xticks(rotation=90)    
    #ax.xaxis.set_ticks_position('left')
    #ax.setp(plt.yticks()[1], rotation=90,fontsize=7)
    ax.set_xlabel(u'Atraso após P(s)',rotation=180,fontsize=9)
    ax.set_ylim(-1,12)
    plt.savefig('Perfil das Funções do Receptor  Sintético.png',dpi=300)
```
