```python
UTM = np.array([utm.from_latlon(i[0],i[1]) for i in zip(lat_x,long_y)])
UTM_E = np.array([int(float(UTM[i][0])) for i in xrange(len(UTM))])
UTM_N = np.array([int(float(UTM[i][1])) for i in xrange(len(UTM))])
```

```python
grid_x, grid_y = np.mgrid([0:1:100j, 0:1:200j])
grid_prof = scipy.interpolate.griddata((UTM_E,UTM_N),prof,(grid_x, grid_y), method = 'cubic')
```

```python
mpl.title("Interpolacao  - Profundidade de Moho")
#mpl.contourf(UTM_E, UTM_N, prof,(10,10),100, interp=True, extrapolate=True)
mpl.plot(UTM_E, UTM_N, '.k')
for label, X, Y in zip(labels, UTM_E, UTM_N):
    mpl.annotate(label,xy=(X,Y), xytext=(-2, 2), ha='right',
                textcoords='offset points',fontsize = 6)
mpl.colorbar()
mpl.xlim(np.min(UTM_E)-50000,np.max(UTM_E)+50000)
mpl.ylim(np.min(UTM_N)-50000,np.max(UTM_N)+50000)
mpl.m2km()
mpl.show()
```
