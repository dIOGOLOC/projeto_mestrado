```python
dados_estacoes = np.genfromtxt('/home/diogo/projeto_mestrado/dados_das_estações.csv',delimiter = ',',skip_header = 1,
                               usecols=(1,2,3,4,5,6,7,8))

labels = np.genfromtxt('/home/diogo/projeto_mestrado/dados_das_estações.csv',delimiter = ',',dtype = "|S7",
                       usecols = 0,skip_header = 1)

lat_x = np.delete(dados_estacoes[:,0],[6,22,23,24])
long_y = np.delete(dados_estacoes[:,1],[6,22,23,24])
prof = np.delete(dados_estacoes[:,3],[6,22,23,24])
labels = np.delete(labels,[6,22,23,24])
delta_prof = np.delete(dados_estacoes[:,4],[6,22,23,24])
vp_vs = np.delete(dados_estacoes[:,5],[6,22,23,24])
delta_vp_vs = np.delete(dados_estacoes[:,6],[6,22,23,24])
numero_FR = np.delete(dados_estacoes[:,7],[6,22,23,24])
labels,prof
```

```python
UTM = np.array([utm.from_latlon(i[0],i[1]) for i in zip(lat_x,long_y)])
UTM_E = np.array([int(float(UTM[i][0])) for i in xrange(len(UTM))])
UTM_N = np.array([int(float(UTM[i][1])) for i in xrange(len(UTM))])
```

```python
areaxy = (np.min(UTM_E)-50000,np.max(UTM_E)+50000,np.min(UTM_N)-50000,np.max(UTM_N)+50000)
shapexy = (100,100)
```

```python
mpl.title("Interpolacao Linear - Profundidade de Moho")
x, y, z = gridder.interp(UTM_E, UTM_N, prof, shapexy, algorithm = 'linear', area = areaxy, extrapolate = True)
mpl.contourf(x, y, z, shapexy, 100, interp = False, extrapolate = False)
mpl.plot(UTM_E, UTM_N, '.k')
for label, X, Y in zip(labels, UTM_E, UTM_N):
    mpl.annotate(label,xy=(X,Y), xytext=(-2, 2), ha='right',
                textcoords='offset points',fontsize = 8)
mpl.colorbar()
mpl.set_area(areaxy)
mpl.m2km()
mpl.show()
```

```python
plt.subplot(111)
mpl.title("Interpolacao Nearest  - Profundidade de Moho")
x, y, z = gridder.interp(UTM_E, UTM_N, prof, shapexy, algorithm = 'nearest', area = areaxy, extrapolate = True)
mpl.contourf(x, y, z, shapexy, 100, interp = False, extrapolate = False)
mpl.plot(UTM_E, UTM_N, '.k')
for label, X, Y in zip(labels, UTM_E, UTM_N):
    mpl.annotate(label,xy=(X,Y), xytext=(-2, 2), ha='right',
                textcoords='offset points',fontsize = 6)
mpl.colorbar()
mpl.set_area(areaxy)
mpl.m2km()
mpl.show()
plt.subplot(111)
mpl.title("Interpolacao Linear  - Profundidade de Moho")
x, y, z = gridder.interp(UTM_E, UTM_N, prof, shapexy, algorithm = 'linear', area = areaxy, extrapolate = True)
mpl.contourf(x, y, z, shapexy, 100, interp = False, extrapolate = False)
mpl.plot(UTM_E, UTM_N, '.k')
for label, X, Y in zip(labels, UTM_E, UTM_N):
    mpl.annotate(label,xy=(X,Y), xytext=(-2, 2), ha='right',
                textcoords='offset points',fontsize = 6)
mpl.colorbar()
mpl.set_area(areaxy)
mpl.m2km()
mpl.show()
mpl.subplot(111)
mpl.title("Interpolacao Cubic  - Profundidade de Moho")
x, y, z = gridder.interp(UTM_E, UTM_N, prof, shapexy, algorithm = 'cubic', area = areaxy, extrapolate = True)
mpl.contourf(x, y, z, shapexy, 100, interp = False, extrapolate = False)
mpl.plot(UTM_E, UTM_N, '.k')
for label, X, Y in zip(labels, UTM_E, UTM_N):
    mpl.annotate(label,xy=(X,Y), xytext=(-2, 2), ha='right',
                textcoords='offset points',fontsize = 6)
mpl.colorbar()
mpl.set_area(areaxy)
mpl.m2km()
mpl.show()
mpl.gcf().set_size_inches(20,20)
mpl.savefig('/home/diogo/projeto_mestrado//Tipos_de_Interpolacao.pdf',dpi=500)
```
