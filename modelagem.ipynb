<h1>Plotando os modelos de Velocidade de Bassini(1986)</h1>

```python
plt.figure(figsize=(6,10))
plt.vlines(5.6,0,-6,'k')
plt.hlines(-6,5.6,6.6,'k')
plt.vlines(6.6,-6,-33,'k')
plt.hlines(-33,6.6,8.1,'k')
plt.vlines(8.1,-33,-60,'k')
plt.xlim(3,9)
plt.xlabel('Velocidade Onda P (km/s)')
plt.ylabel('Profundidade (km)')
```

```python
plt.figure(figsize=(10,5))
plt.plot(b.times()-5,b.data,'k',linewidth=1)
plt.fill_between(b.times()-5,b.data,0,where=b.data>=0,facecolor='black',alpha=0.7, interpolate=True)
plt.fill_between(b.times()-5,b.data,0,where=b.data<=0, facecolor='gray',alpha=0.7, interpolate=True)
plt.yticks([])
plt.xlim(-3,35)
```
